```python
# Import dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Read in the CSV file as a Pandas DataFrame
data_df = pd.read_csv(
    Path("data/Most-Recent-Cohorts-Institution.csv")
)

# Review the DataFrame
data_df.head(3)
```

```python
#select columns to keeps
desired_columns = ['UNITID', 'STABBR', 'HIGHDEG','CONTROL', 'COSTT4_A', 'TUITIONFEE_IN', 'TUITIONFEE_OUT', 'GRAD_DEBT_MDN10YR', 'GRAD_DEBT_MDN_SUPP', 'NPT4_PUB', 'MN_EARN_WNE_P10', 'MN_EARN_WNE_P6', 'MN_EARN_WNE_P7', 'MN_EARN_WNE_P8', 'MN_EARN_WNE_P9'   
                ]
only_cols = data_df.loc[:, desired_columns]


#data_df[desired_columns] = data_df[desired_columns].replace('PrivacySuppressed', 'NaN')
only_cols.head(50)
```

```python
column_names = ['UNITID','STABBR', 'HIGHDEG','CONTROL', 'COSTT4_A', 'TUITIONFEE_IN', 'TUITIONFEE_OUT', 'GRAD_DEBT_MDN10YR', 'GRAD_DEBT_MDN_SUPP', 'NPT4_PUB', 'MN_EARN_WNE_P10', 'MN_EARN_WNE_P6', 'MN_EARN_WNE_P7', 'MN_EARN_WNE_P8', 'MN_EARN_WNE_P9' ]
cleaned_df = only_cols.dropna(subset= column_names)
cleaned_df.head(50)
```

```python
cleaned_df = cleaned_df[(cleaned_df['GRAD_DEBT_MDN10YR'] != 'PrivacySuppressed') & 
                    (cleaned_df['GRAD_DEBT_MDN_SUPP'] != 'PrivacySuppressed') & 
                    (cleaned_df['MN_EARN_WNE_P10'] != 'PrivacySuppressed') & 
                    (cleaned_df['MN_EARN_WNE_P6'] != 'PrivacySuppressed') & 
                    (cleaned_df['MN_EARN_WNE_P7'] != 'PrivacySuppressed') & 
                    (cleaned_df['MN_EARN_WNE_P8'] != 'PrivacySuppressed') & 
                    (cleaned_df['MN_EARN_WNE_P9'] != 'PrivacySuppressed')]
cleaned_df.head()
```

```python
#change data type from object to float
cleaned_df = cleaned_df.astype({"COSTT4_A":"int","TUITIONFEE_IN":"int", "TUITIONFEE_OUT":"int","GRAD_DEBT_MDN10YR":"float","GRAD_DEBT_MDN_SUPP":"int","NPT4_PUB":"int", "MN_EARN_WNE_P10":"int", "MN_EARN_WNE_P6":"int",
                        "MN_EARN_WNE_P7":"int", "MN_EARN_WNE_P8":"int", "MN_EARN_WNE_P9":"int"})
cleaned_df.info()
```

```python
cleaned_df.to_csv("data/Clean_Most-Recent-Cohorts-Institution.csv")
```

```python
data_df = pd.read_csv(
    Path("data/Most-Recent-Cohorts-Field-of-Study.csv")
)

# Review the DataFrame
data_df.head(20)
```

```python
#select columns to keep

study_columns= ['UNITID', 'INSTNM', 'CIPCODE', 'CREDLEV', 'CIPDESC','DEBT_ALL_STGP_ANY_MEAN', 'DEBT_ALL_PP_ANY_MEAN','DEBT_ALL_PP_EVAL_MEAN','DEBT_ALL_STGP_EVAL_MEAN', 'DEBT_ALL_PP_ANY_MDN10YRPAY', 'DEBT_ALL_PP_EVAL_MDN10YRPAY', 
               'EARN_MDN_HI_1YR', 'EARN_MDN_HI_2YR', 'EARN_NE_MDN_3YR', 'EARN_MDN_4YR']
these_cols = data_df.loc[:, study_columns]


these_cols.head(50)
```

```python
#drop NA values
column_nms = ['UNITID', 'CIPCODE', 'CREDLEV', 'DEBT_ALL_STGP_ANY_MEAN', 'DEBT_ALL_PP_ANY_MEAN','DEBT_ALL_PP_EVAL_MEAN','DEBT_ALL_STGP_EVAL_MEAN', 'DEBT_ALL_PP_ANY_MDN10YRPAY', 'DEBT_ALL_PP_EVAL_MDN10YRPAY', 
               'EARN_MDN_HI_1YR', 'EARN_MDN_HI_2YR', 'EARN_NE_MDN_3YR', 'EARN_MDN_4YR']

# Initialize the DataFrame 'cleaned_df' as a copy of 'only_cols'
clean_df = these_cols.copy()

# Iterate through the specified columns and drop rows with NaN values in each column
for col in column_nms:
    clean_df = clean_df[clean_df[col].notnull()]

clean_df.head()
```

```python
clean_df = clean_df[(clean_df['DEBT_ALL_STGP_ANY_MEAN'] != 'PrivacySuppressed') & 
                    (clean_df['DEBT_ALL_PP_ANY_MEAN'] != 'PrivacySuppressed') & 
                    (clean_df['DEBT_ALL_PP_EVAL_MEAN'] != 'PrivacySuppressed') & 
                    (clean_df['DEBT_ALL_STGP_EVAL_MEAN'] != 'PrivacySuppressed') & 
                    (clean_df['DEBT_ALL_PP_ANY_MDN10YRPAY'] != 'PrivacySuppressed') & 
                    (clean_df['DEBT_ALL_PP_EVAL_MDN10YRPAY'] != 'PrivacySuppressed') & 
                    (clean_df['EARN_MDN_HI_1YR'] != 'PrivacySuppressed') & 
                    (clean_df['EARN_MDN_HI_2YR'] != 'PrivacySuppressed') & 
                    (clean_df['EARN_NE_MDN_3YR'] != 'PrivacySuppressed')& 
                    (clean_df['EARN_MDN_4YR'] != 'PrivacySuppressed')]
clean_df.head()
```

```python
clean_df.info()
```

```python
clean_df = clean_df.astype({"UNITID":"int", "DEBT_ALL_STGP_ANY_MEAN":"int","DEBT_ALL_PP_ANY_MEAN":"float", "DEBT_ALL_PP_EVAL_MEAN":"float", "DEBT_ALL_STGP_EVAL_MEAN":"float",
                        "DEBT_ALL_PP_ANY_MDN10YRPAY":"float", "DEBT_ALL_PP_EVAL_MDN10YRPAY":"float", "EARN_MDN_HI_1YR":"float", "EARN_MDN_HI_2YR":"float", "EARN_NE_MDN_3YR":"float", "EARN_MDN_4YR":"float"})
```
